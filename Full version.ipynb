Patterns for Transactions Fields

```python
# Transaction Date
patt_date = r'(?<=:32A:)\d{6}' # takes the first six characters

match_date = re.search(patt_date, message)
if match_date:
    date_str = match_date.group(0)
    trans_date = date_str
    print(f"Transaction date extracted from message: {trans_date}")
else:
    print("No date found in message.")

# Transaction ID
patt_transid = r':20:(.*)\n'
match = re.search(patt_transid, message)
if match:
    transaction_id = match.group(1)
    trans_id = transaction_id
    print(f"Transaction ID extracted from message: {trans_id}")
else:
    print("No transaction ID found in message.")

# Transaction Message


# Transaction Currency
patt_curr = r'(?<=:32A:\d{6})[A-Z]{3}' # takes the first three characters after the sixth character

match_curr = re.search(patt_curr, message)
if match_curr:
    curr_str = match_curr.group(0)
    trans_curr = curr_str
    print(f"Currency extracted from message: {trans_curr}")
else:
    print("No currency found in text.")

# Transaction Amount 
patt_amt = r'(?<=:32A:\d{6}[A-Z]{3})\d+([.,]\d+)?' 

match_amt = re.search(patt_amt, message)
if match_amt:
    amt_str = match_amt.group(0)
    amt_float = float(amt_str.replace(',', '.'))
    trans_amt = amt_float
    print(f"Amount extracted from message: {trans_amt}")
else:
    print("No amount found in message.")


# Transaction Type
patt_typ = r'(?<=:23B:)[A-Z]+'

match_typ = re.search(patt_typ, message)

if match_typ:
    trans_type = match_typ.group(0)
    print('Transaction type is:', trans_type)
else:
    print('Transaction type not found.')

# patt_dir = any("i", "o", "io")
# patt_stat = any("accepted", "rejected")

# patt_ins_typ = any("cash","check","ach/lcy_transfers","wire","securities","e-
# money/mobile_money","travellers_cheques","prepaid_cards","certified_ch
# eques","vouchers","cashier_cheques/money_order","precious_metal","cryp
# to/virtual_assets","interest/dividend","other")
```

Patterns for Originator Fields

```python
# Full name (needs checking)
patt_ori_full = r":50[AFK]:\s*\/.*\n(.*)"

    # Extracts the text on the first line after :50K:
match_ori_full = re.search(patt_ori_full, message)
if match_ori_full:
    originator_fullname = match_ori_full.group(1)
    print('originator full name is:', originator_fullname)
else:
    print('No match found for originator full name.')

# Originator Address = city
patt_ori_cty = r":50[AFK]:\s*\/.*\n.*\n(.*),"

    # Extract the text before the comma in the third line after :50K:
match_ori_cty = re.search(patt_ori_cty, message)
if match_ori_cty:
    ori_city = match_ori_cty.group(1)
    print('originator address is:', ori_city)
else:
    print('No match found for originator address.')

# Originator Country
patt_ori_cnty = r":50[AFK]:\s*\/.*\n.*\n.*, (.*)"

    # Extract the text after the comma in the third line after :50K:
match_ori_cnty = re.search(patt_ori_cnty, message)
if match_ori_cnty:
    ori_cnty = match_ori_cnty.group(1)
    print('originator country is:', ori_cnty)
else:
    print('No match found for originator country.')

# Originator Account Number
patt_ori_actnr = r'(?<=\n):50[AFK]:(.+)'
matches_actnr = re.findall(patt_ori_actnr, message)

    # Extract account number after :50[AFK]: and removing '/'
for match_actnr in matches_actnr:
    match_actnr = match_actnr.strip()
    if match_actnr.startswith('/'):
        match_actnr = match_actnr[1:]
        ori_actnr = match_actnr
        print('originator account number is:', ori_actnr)

# Originator Bank Identifier Code
patt_ori_bic = r'(?<=\n):52[AD]:(.+)'

match_bic = re.search(patt_ori_bic, message)
if match_bic:
    ori_bic = match_bic.group(1)
    ori_bic = ori_bic.strip().replace("/", "")
    print('originator BIC is:', ori_bic)
```

Patterns for Intermediaries

```python
# Receiving Intermediary Financial Institution BIC
patt_rec_bic = r'(?<=\n):56[ABD]:(.+)'

match_rec_bic = re.search(patt_rec_bic, message)
if match_rec_bic:
    rec_bic = match_rec_bic.group(1)
    rec_bic = rec_bic.strip().replace("/", "")
    print('receiving intermediary FI BIC is:', rec_bic)
else:
    print('No receiving intermediary FI BIC found.')

# Sending Intermediary Financial Institution BIC
patt_send_bic = r'(?<=\n):57[ABD]:(.+)'

match_send_bic = re.search(patt_send_bic, message)
if match_send_bic:
    send_bic = match_send_bic.group(1)
    send_bic = send_bic.strip().replace("/", "")
    print('sending intermediary FI BIC is:', send_bic)
```

```python
# Transaction Date
patt_date = r'(?<=:32A:)\d{6}' # takes the first six characters

match_date = re.search(patt_date, message3)
if match_date:
    date_str = match_date.group(0)
    trans_date = date_str
    print(f"Transaction date extracted from message: {trans_date}")
else:
    print("No date found in message.")

# Transaction ID
patt_transid = r':20:(.*)\n'
match = re.search(patt_transid, message3)
if match:
    transaction_id = match.group(1)
    trans_id = transaction_id
    print(f"Transaction ID extracted from message: {trans_id}")
else:
    print("No transaction ID found in message.")

# Transaction Message


# Transaction Currency
patt_curr = r'(?<=:32A:\d{6})[A-Z]{3}' # takes the first three characters after the sixth character

match_curr = re.search(patt_curr, message3)
if match_curr:
    curr_str = match_curr.group(0)
    trans_curr = curr_str
    print(f"Currency extracted from message: {trans_curr}")
else:
    print("No currency found in text.")

# Transaction Amount 
patt_amt = r'(?<=:32A:\d{6}[A-Z]{3})\d+([.,]\d+)?' 

match_amt = re.search(patt_amt, message3)
if match_amt:
    amt_str = match_amt.group(0)
    amt_float = float(amt_str.replace(',', '.'))
    trans_amt = amt_float
    print(f"Amount extracted from message: {trans_amt}")
else:
    print("No amount found in message.")


# Transaction Type
patt_typ = r'(?<=:23B:)[A-Z]+'

match_typ = re.search(patt_typ, message3)

if match_typ:
    trans_type = match_typ.group(0)
    print('Transaction type is:', trans_type)
else:
    print('Transaction type not found.')

# patt_dir = any("i", "o", "io")
# patt_stat = any("accepted", "rejected")

# patt_ins_typ = any("cash","check","ach/lcy_transfers","wire","securities","e-
# money/mobile_money","travellers_cheques","prepaid_cards","certified_ch
# eques","vouchers","cashier_cheques/money_order","precious_metal","cryp
# to/virtual_assets","interest/dividend","other")
```

```python
# Full name (needs checking)
patt_ori_full = r":50[AFK]:\s*\/.*\n(.*)"

    # Extracts the text on the first line after :50K:
match_ori_full = re.search(patt_ori_full, message3)
if match_ori_full:
    originator_fullname = match_ori_full.group(1)
    print('originator full name is:', originator_fullname)
else:
    print('No match found for originator full name.')

# Originator Address = city
patt_ori_cty = r":50[AFK]:\s*\/.*\n.*\n(.*),"

    # Extract the text before the comma in the third line after :50K:
match_ori_cty = re.search(patt_ori_cty, message3)
if match_ori_cty:
    ori_city = match_ori_cty.group(1)
    print('originator address is:', ori_city)
else:
    print('No match found for originator address.')

# Originator Country
patt_ori_cnty = r":50[AFK]:\s*\/.*\n.*\n.*, (.*)"

    # Extract the text after the comma in the third line after :50K:
match_ori_cnty = re.search(patt_ori_cnty, message3)
if match_ori_cnty:
    ori_cnty = match_ori_cnty.group(1)
    print('originator country is:', ori_cnty)
else:
    print('No match found for originator country.')

# Originator Account Number
patt_ori_actnr = r'(?<=\n):50[AFK]:(.+)'
matches_actnr = re.findall(patt_ori_actnr, message3)

    # Extract account number after :50[AFK]: and removing '/'
for match_actnr in matches_actnr:
    match_actnr = match_actnr.strip()
    if match_actnr.startswith('/'):
        match_actnr = match_actnr[1:]
        ori_actnr = match_actnr
        print('originator account number is:', ori_actnr)

# Originator Bank Identifier Code
patt_ori_bic = r'(?<=\n):52[AD]:(.+)'

match_bic = re.search(patt_ori_bic, message3)
if match_bic:
    ori_bic = match_bic.group(1)
    ori_bic = ori_bic.strip().replace("/", "")
    print('originator BIC is:', ori_bic)
```

```python
# Receiving Intermediary Financial Institution BIC
patt_rec_bic = r'(?<=\n):56[ABD]:(.+)'

match_rec_bic = re.search(patt_rec_bic, message3)
if match_rec_bic:
    rec_bic = match_rec_bic.group(1)
    rec_bic = rec_bic.strip().replace("/", "")
    print('receiving intermediary FI BIC is:', rec_bic)
else:
    print('No receiving intermediary FI BIC found.')

# Sending Intermediary Financial Institution BIC
patt_send_bic = r'(?<=\n):57[ABD]:(.+)'

match_send_bic = re.search(patt_send_bic, message3)
if match_send_bic:
    send_bic = match_send_bic.group(1)
    send_bic = send_bic.strip().replace("/", "")
    print('sending intermediary FI BIC is:', send_bic)
```
